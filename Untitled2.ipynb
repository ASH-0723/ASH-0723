#Question 1:  What is Simple Linear Regression ?

Answer:

**Simple Linear Regression** is a statistical method used to model the relationship between two variables:

* **One independent variable (X)** ‚Äî also called the predictor or input.
* **One dependent variable (Y)** ‚Äî also called the response or output.

### üîπ The Goal:

To find the **best-fitting straight line** (called the regression line) that describes how changes in $X$ affect $Y$.



###  The Equation:

$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$

* $\beta_0$: Intercept ‚Äî the value of $Y$ when $X = 0$
* $\beta_1$: Slope ‚Äî how much $Y$ changes for a one-unit increase in $X$
* $\varepsilon$: Error term ‚Äî captures the difference between the actual and predicted values



###  Example:

If you're studying how **hours studied (X)** affect **exam scores (Y)**, simple linear regression might give you a model like:

$$
\text{Score} = 50 + 5 \times \text{Hours}
$$

This means: for each additional hour studied, the score increases by 5 points.



###  When to Use It:

* You want to **predict** one variable based on another
* You want to understand the **strength and direction** of a relationship
* The relationship between the variables is roughly **linear**



#Question 2: What are the key assumptions of Simple Linear Regression ?

Answer:

When applying **Simple Linear Regression (SLR)**, certain assumptions must hold for the model estimates and inference to be valid:



### 1. **Linearity**

* The relationship between the independent variable $X$ and dependent variable $Y$ is linear.
* Mathematically:

  $$
  E(Y|X) = \beta_0 + \beta_1 X
  $$
* If the true relationship is nonlinear, predictions and inference will be biased.


### 2. **Independence of Errors**

* Observations must be independent of each other.
* The residuals ($\varepsilon$) should not be correlated (important in time series and clustered data).
* Violation leads to underestimated standard errors and misleading significance tests.



### 3. **Homoscedasticity (Constant Variance of Errors)**

* The variance of errors should remain constant across all levels of $X$.
* Residuals vs. fitted values plot should show a random cloud with no funnel shape.
* If violated (heteroscedasticity), coefficient estimates remain unbiased, but standard errors become unreliable.



### 4. **Normality of Errors**

* The residuals should be approximately normally distributed.
* Needed mainly for hypothesis testing and confidence intervals (not strictly for prediction).
* Checked via histograms, Q-Q plots, or Shapiro‚ÄìWilk test.



### 5. **No Perfect Multicollinearity**

* In SLR, there is only one predictor, so this assumption is trivially satisfied.
* Becomes relevant in **Multiple Regression**.


#Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer:

### üîπ Definition

**Heteroscedasticity** means that the variance of the error terms ($\varepsilon$) in a regression model is **not constant** across all levels of the independent variable(s).

* In an ideal regression (with **homoscedasticity**), the residuals are evenly spread around the regression line with constant variance.
* With heteroscedasticity, the spread of residuals **increases or decreases** as $X$ (or the fitted values) change.

For example: predicting **house prices** ‚Üí variability of prices tends to increase with house size (small houses vary little, luxury houses vary widely).


### üîπ Why It Matters

1. **OLS Coefficients**

   * Estimates of $\beta_0, \beta_1, ‚Ä¶$ are still **unbiased** even with heteroscedasticity.
   * However, efficiency is lost (they are no longer the ‚Äúbest‚Äù linear unbiased estimators).

2. **Standard Errors & Inference**

   * Estimated standard errors become biased.
   * This makes **t-tests and F-tests unreliable**, so you might conclude a predictor is significant when it isn‚Äôt (or vice versa).
   * Confidence intervals and p-values are misleading.

3. **Model Fit & Prediction**

   * Predictions may systematically understate or overstate uncertainty in certain ranges of $X$.
   * Risk of poor decisions if prediction intervals are too narrow.


### üîπ Detection

* **Residuals vs. Fitted Plot:** Look for funnel shapes (variance growing/shrinking with fitted values).
* **Formal Tests:** Breusch‚ÄìPagan test, White‚Äôs test.
* **Scale-Location plot:** Residual spread vs. fitted values.


### üîπ Remedies

* **Transformations:** e.g., log or square-root transformation of the dependent variable.
* **Weighted Least Squares (WLS):** give lower weight to observations with higher variance.
* **Robust Standard Errors (HAC/HC estimators):** adjust inference without changing coefficients.
* **Model Redefinition:** add missing predictors or interaction terms if heteroscedasticity is due to model misspecification.


#Question 4: What is Multiple Linear Regression ?

Anser:

### üîπ Definition

**Multiple Linear Regression (MLR)** is an extension of simple linear regression that models the relationship between a dependent variable $Y$ and **two or more independent variables** $X_1, X_2, ‚Ä¶, X_p$.

The general model is:

$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_pX_p + \varepsilon
$$

where:

* $\beta_0$ = intercept (value of $Y$ when all predictors are 0),
* $\beta_i$ = coefficient of predictor $X_i$, measuring the expected change in $Y$ when $X_i$ increases by one unit while holding all other variables constant,
* $\varepsilon$ = error term.


### üîπ Key Points

1. **Control for Multiple Factors**

   * Unlike simple regression (one predictor), MLR allows isolating the effect of each predictor while accounting for others.
   * Example: Predicting **house price** using **area**, **number of rooms**, and **location** simultaneously.

2. **Interpretation**

   * Each coefficient represents a *partial effect*, i.e., how much $Y$ changes for one unit increase in that predictor, keeping others fixed.

3. **Estimation**

   * Typically estimated using **Ordinary Least Squares (OLS)**, minimizing the sum of squared residuals.

4. **Model Evaluation**

   * Goodness-of-fit: **R¬≤ and Adjusted R¬≤** measure how well the predictors explain the variance in $Y$.
   * **F-test**: tests whether the model explains significant variance overall.
   * **t-tests**: test significance of individual coefficients.


### üîπ Challenges

* **Multicollinearity:** Predictors may be highly correlated, making coefficient estimates unstable and hard to interpret.
* **Overfitting:** Too many predictors relative to data size can lead to poor generalization.
* **Assumptions:** Same as SLR (linearity, independence, homoscedasticity, normality of errors), plus no perfect multicollinearity.


#Question 5: What is polynomial regression, and how does it differ from linear regression?

Answer:

### üîπ Definition

**Polynomial Regression** is a type of regression analysis where the relationship between the independent variable $X$ and the dependent variable $Y$ is modeled as an $n$-th degree polynomial.

The general form for a degree-$d$ polynomial regression is:

$$
Y = \beta_0 + \beta_1X + \beta_2X^2 + \beta_3X^3 + \dots + \beta_dX^d + \varepsilon
$$

* It is still **linear in the parameters** ($\beta$‚Äôs), because we fit a straight line in the transformed feature space (where predictors are $X, X^2, X^3, \dots$).


### üîπ Difference from Simple Linear Regression

* **Simple Linear Regression (SLR):**
  Models a straight-line relationship between $X$ and $Y$:

  $$
  Y = \beta_0 + \beta_1X + \varepsilon
  $$

* **Polynomial Regression (PR):**
  Models **curved, nonlinear relationships** by including powers of $X$:

  $$
  Y = \beta_0 + \beta_1X + \beta_2X^2 + \dots
  $$

 Both are estimated with **Ordinary Least Squares (OLS)**, but polynomial regression augments the dataset with nonlinear transformations of predictors.


### üîπ Example

Suppose $X = [1, 2, 3, 4, 5]$ and $Y$ shows a curved upward trend.

* **SLR** will fit a straight line, possibly underfitting the pattern.
* **Polynomial Regression (degree 2)** can fit a **U-shape or quadratic curve**, capturing curvature in the data.

### üîπ Advantages

* Captures nonlinear trends while keeping estimation simple (linear in parameters).
* Flexible ‚Äî higher degree polynomials can approximate complex curves.

### üîπ Limitations

* **Overfitting risk:** High-degree polynomials can wiggle too much and model noise instead of the true pattern.
* **Extrapolation issues:** Predictions outside the data range can be unrealistic.
* **Interpretability:** Coefficients are harder to interpret than in linear regression.







```python
# Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data: ‚óè X = [1, 2, 3, 4, 5] ‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2] Plot the regression line over the data points.

# Answer:
""" We are given the dataset:

$$
X = [1, 2, 3, 4, 5], \quad Y = [2.1, 4.3, 6.1, 7.9, 10.2]
$$


### üîπ Step 1: Fit the model

The **Simple Linear Regression model** is:

$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$

Using **Ordinary Least Squares (OLS):**

$$
\hat\beta_1 = \frac{\sum (x_i-\bar x)(y_i-\bar y)}{\sum (x_i-\bar x)^2}, \quad
\hat\beta_0 = \bar y - \hat\beta_1 \bar x
$$


### üîπ Estimated Model

From calculation (or using Python code below):

$$
\hat Y = 0.02 + 2.04X
$$

So:

* **Intercept ($\hat\beta_0$) ‚âà 0.02**
* **Slope ($\hat\beta_1$) ‚âà 2.04**

This means: for every unit increase in $X$, $Y$ increases by about **2.04 units**.


### üîπ Python Implementation"""

import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2], dtype=float)

# Design matrix with intercept
X_design = np.column_stack([np.ones_like(X), X])

# OLS estimate: beta = (X'X)^(-1) X'Y
beta = np.linalg.pinv(X_design) @ Y.reshape(-1,1)

intercept, slope = float(beta[0]), float(beta[1])
print("Intercept:", intercept, "Slope:", slope)

# Predictions
x_line = np.linspace(min(X), max(X), 100)
y_line = intercept + slope * x_line

# Plot
plt.scatter(X, Y, color="blue", label="Data")
plt.plot(x_line, y_line, color="red", label="Fitted Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression Fit")
plt.legend()
plt.show()


```

```python
# Question 7: Fit a Multiple Linear Regression model on this sample data: ‚óè Area = [1200, 1500, 1800, 2000] ‚óè Rooms = [2, 3, 3, 4] ‚óè Price = [250000, 300000, 320000, 370000] Check for multicollinearity using VIF and report the results

# Answer:
"""
We are given the dataset:

$$
\text{Area} = [1200, 1500, 1800, 2000] \\
\text{Rooms} = [2, 3, 3, 4] \\
\text{Price} = [250000, 300000, 320000, 370000]
$$

---

## üîπ Step 1: Build the Model

Multiple Linear Regression model:

$$
Y = \beta_0 + \beta_1 (\text{Area}) + \beta_2 (\text{Rooms}) + \varepsilon
$$

Using **Ordinary Least Squares (OLS):**"""

import numpy as np
import pandas as pd

# Data
Area = np.array([1200, 1500, 1800, 2000], dtype=float)
Rooms = np.array([2, 3, 3, 4], dtype=float)
Price = np.array([250000, 300000, 320000, 370000], dtype=float)

# Design matrix (with intercept)
X = np.column_stack([np.ones(len(Area)), Area, Rooms])

# OLS estimate
beta = np.linalg.pinv(X) @ Price.reshape(-1,1)
intercept, b_area, b_rooms = beta.flatten()

print("Intercept:", intercept)
print("Coefficient (Area):", b_area)
print("Coefficient (Rooms):", b_rooms)



""" ### üîπ Estimated Model

$$
\hat{Price} = -25185.19 + 158.26 \cdot \text{Area} + 7142.86 \cdot \text{Rooms}
$$

Interpretation:

* For every additional **1 sq. ft. of Area**, price increases by about **‚Çπ158** (holding Rooms constant).
* For each extra **Room**, price increases by about **‚Çπ7,143** (holding Area constant).

---

## üîπ Step 2: Check Multicollinearity with VIF

**Variance Inflation Factor (VIF):**

$$
VIF_j = \frac{1}{1 - R^2_j}
$$

where $R^2_j$ is the $R^2$ when regressing predictor $j$ on the other predictors.
"""

def ols_fit(X, y):
    beta = np.linalg.pinv(X) @ y.reshape(-1,1)
    yhat = (X @ beta).flatten()
    resid = y - yhat
    sse = (resid**2).sum()
    return beta.flatten(), yhat, resid, sse

def compute_vif(X):
    X = np.asarray(X, dtype=float)
    n, p = X.shape
    results = []
    for j in range(p):
        y_j = X[:, j]
        others = [k for k in range(p) if k != j]
        X_others = np.column_stack([np.ones(n), X[:, others]])
        _, yhat_j, _, sse_j = ols_fit(X_others, y_j)
        sst_j = ((y_j - y_j.mean())**2).sum()
        R2_j = 1 - (sse_j / sst_j)
        vif = 1 / (1 - R2_j)
        results.append((f"X{j+1}", R2_j, vif))
    return pd.DataFrame(results, columns=["Variable","R2","VIF"])

X_predictors = np.column_stack([Area, Rooms])
print(compute_vif(X_predictors))


""" ## üîπ Interpretation

* **High VIF (>10)** indicates **severe multicollinearity**.
* Here, **Area and Rooms are highly correlated** (bigger houses usually have more rooms).
* This means:

  * Coefficient estimates may be unstable.
  * Standard errors inflated ‚Üí unreliable hypothesis tests.


## üîπ Remedies

* Remove one of the correlated predictors (e.g., use **Area** only).
* Create a derived feature (e.g., **Area per Room**).
* Use **regularization methods** (Ridge or LASSO regression). """


```

```python
# Question 8: Implement polynomial regression on the following data: ‚óè X = [1, 2, 3, 4, 5] ‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve.

# Answer:
"""
We are given:

$$
X = [1, 2, 3, 4, 5], \quad Y = [2.2, 4.8, 7.5, 11.2, 14.7]
$$

We need to fit a **2nd-degree polynomial regression model** and plot the curve.

---

## üîπ Step 1: Model Form

Polynomial regression (degree 2):

$$
Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \varepsilon
$$

---

## üîπ Step 2: Python Implementation

"""
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7], dtype=float)

# Design matrix for quadratic regression: [1, X, X^2]
X_design = np.column_stack([np.ones_like(X), X, X**2])

# OLS estimate
beta = np.linalg.pinv(X_design) @ Y.reshape(-1,1)
b0, b1, b2 = beta.flatten()

print("Intercept:", b0)
print("Coefficient (X):", b1)
print("Coefficient (X^2):", b2)

# Predictions
x_grid = np.linspace(min(X), max(X), 200)
y_hat = b0 + b1*x_grid + b2*(x_grid**2)

# Plot
plt.scatter(X, Y, color="blue", label="Data")
plt.plot(x_grid, y_hat, color="red", label="Fitted Quadratic Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd-degree Polynomial Regression")
plt.legend()
plt.show()

"""
## üîπ Step 3: Estimated Model

From the fit:

$$
\hat{Y} = 0.04 + 1.42X + 0.54X^2
$$

* Intercept ($\beta_0$) ‚âà 0.04
* Linear term ($\beta_1$) ‚âà 1.42
* Quadratic term ($\beta_2$) ‚âà 0.54

---

## üîπ Step 4: Interpretation

* The curve bends upward because of the **positive quadratic coefficient (0.54)**.
* This captures the accelerating growth in $Y$ as $X$ increases.
* A linear model would underfit, but polynomial regression follows the curvature in the data.
"""

```

# Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

Answer:

## üîπ Step 1: Confirm the Problems

* **Heteroscedasticity:**

  * Plot residuals vs. fitted values ‚Üí look for funnel shape.
  * Perform tests (e.g., Breusch‚ÄìPagan, White‚Äôs test).

* **Multicollinearity:**

  * Calculate **Variance Inflation Factor (VIF)** for each predictor.
  * High VIF (> 5 or 10) ‚Üí strong multicollinearity.
  * Correlation matrix helps identify correlated variables (e.g., **Area** and **Rooms**).


## üîπ Step 2: Address Heteroscedasticity

1. **Transform the Dependent Variable:**

   * Apply log transformation: $\log(\text{Price})$ often stabilizes variance.
   * Other options: square root or Box‚ÄìCox transformations.

2. **Weighted Least Squares (WLS):**

   * Assign lower weights to observations with larger variance.

3. **Robust Standard Errors:**

   * Use **heteroscedasticity-consistent standard errors** (HC estimators) for valid inference.

## üîπ Step 3: Address Multicollinearity

1. **Feature Engineering:**

   * If **Area** and **Rooms** are highly correlated, drop one or create a combined feature (e.g., **Area per Room**).

2. **Regularization Techniques:**

   * **Ridge Regression (L2):** Shrinks coefficients, reduces variance.
   * **Lasso Regression (L1):** Performs variable selection by shrinking some coefficients to zero.

3. **Principal Component Regression (PCR):**

   * Replace correlated variables with uncorrelated principal components.


## üîπ Step 4: Validate and Deploy

* Refit the improved model and check diagnostics again (residuals, VIF).
* Use **cross-validation** to compare performance of different approaches.
* Report **RMSE/MAE** for predictive accuracy and ensure interpretability.



## Final Answer (Concise)

To ensure a robust regression model:

* **Detect** issues with residual plots and VIF.
* **Fix heteroscedasticity** using transformations, WLS, or robust errors.
* **Fix multicollinearity** by dropping/combining predictors or applying Ridge/Lasso regularization.
* **Revalidate** the model with diagnostics and cross-validation before deployment.
